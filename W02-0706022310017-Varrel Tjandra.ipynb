```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url, sep=',')

# 2
df.head()
```

```python
df.info()
# Diketahui terdapat 2 type data yaitu object dan float
# Terdapat missing value pada class data no 9 - 14, 16, 20-24, dan 26
```

```python
df.describe()
# kalau dilihat sekilas terdapat perbedaan outlier yang sangat jauh nilai nya pada sebagian besar variable seperti kamar tidur, kamar mandi, luas tanah
# garasi dan lain lain yang dapat dilihat dari median seperti garasi yang mediannya 0 yang berarti sebagian besar property tidak memiliki garasi.
```

```python
# 3
# Mencari tahu sebeapa banyak missing value dan presentase missing valuenya

# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
# Mengubah value electricity yang berisikan "lainnya mah" menjadi unknown karena kalau diubah menjadi sebuah median atau mean nantinya tidak relevan dengan keadaan yang asli
# Mengubah property_type yang null menjadi "rumah" karena jika dilihat dari unique value hanya terdapat rumah, oleh karena itu saya mengubah menjadi rumah.

# Replace "lainnya mah" with "unknown" in the 'electricity' column
df['electricity'] = df['electricity'].replace("lainnya mah", "unknown")

# Replace null values in 'property_type' with 'rumah'
df['property_type'] = df['property_type'].fillna('rumah')

# Verify the changes
print("Unique values in 'electricity' after replacement:")
print(df['electricity'].unique())
print("\nNull values in 'property_type' after imputation:")
print(df['property_type'].isnull().sum())
```

```python
# Mengubah data yang null menjadi "unknown" karena jika nanti data yang null semisal diisi oleh modus, maka bisa menyebabkan terjadi
# kesalahpahaman. Semisal property_type yang aslinya adalah rumah sakit namun ketika diisi modus menjadi rumah. Untuk menghindari hal
# Tersebut maka saya mengubah menjadi unknown supaya dapat dipisahkan dan dapat diupdate kedepannya

# Proses mengganti data yang null dengan data type object menjadi unknown
object_cols = df.select_dtypes(include=['object']).columns

# Ganti null ke unknown
for col in object_cols:
    df[col] = df[col].fillna("unknown")

print("Null values setelah imputation:")
print(df[object_cols].isnull().sum())
```

```python
# Menentukan kolom mana yang terdapat null dan cocok untuk imputasi rata-rata atau median.

numerical_cols_with_nulls = null_df.index[null_df.index.isin(df.select_dtypes(include=['float64', 'int64']).columns)].tolist()
print("Numerical columns with missing values suitable for mean or median imputation:")
print(numerical_cols_with_nulls)
```

```python
# Menghitung skewness untuk menentukan agar memakai median atau mean untuk mengisi null value
# Skewness = 0 â†’ distribusi relatif simetris (mendekati normal).
# Skewness > 0 (positif) â†’ distribusi condong ke kanan (right-skewed). Banyak data kecil, beberapa nilai ekstrem besar.
# Skewness < 0 (negatif) â†’ distribusi condong ke kiri (left-skewed). Banyak data besar, beberapa nilai ekstrem kecil.

for col in numerical_cols_with_nulls:
  skewness = df[col].skew()
  print(f"Skewness for column '{col}': {skewness}")
```

```python
# Menentukan memakai median atau mean dalam imputasi
# Jika skewness diantara -1 dan 1, maka akan memakai median karena mendekati normal sedangkan jika lebih atau kurang dari itu akan memakai median

imputation_strategies = {}
skewness_threshold = 1.0  # Define a skewness threshold for determining skewness

for col in numerical_cols_with_nulls:
  skewness = df[col].skew()
  if abs(skewness) > skewness_threshold:
    imputation_strategies[col] = 'median'
    print(f"Column '{col}' is skewed (skewness = {skewness:.2f}), imputing with median.")
  else:
    imputation_strategies[col] = 'mean'
    print(f"Column '{col}' is symmetrical (skewness = {skewness:.2f}), imputing with mean.")

print("\nImputation Strategies:")
print(imputation_strategies)
```

```python
# Mencari kolom yang dapat digunakan untuk mengelompokkan data untuk imputasi yang lebih akurat
# (misalnya, imputasi building_age berdasarkan kota atau distrik).

print("Unique values for column 'city':")
print(df['city'].unique())
print("-" * 20)

print("Unique values for column 'district':")
print(df['district'].unique())
print("-" * 20)

print("Unique values for column 'property_type':")
print(df['property_type'].unique())
print("-" * 20)
```

```python
# Berdasarkan nilai uniknya, 'kota' dan 'distrik'dapat dikelompokkan karena bervariasi.
# 'Property_type' juga memiliki kategori berbeda yang dapat memengaruhi karakteristik properti.
grouping_columns = ['city', 'district', 'property_type']
print("Selected columns for potential grouping:")
print(grouping_columns)
```

```python
# Menerapkan imputasi yang dipilih (mean atau median) untuk mengisi nilai yang hilang di setiap kolom yang relevan

grouping_columns = ['city', 'district'] # Grouping columns based on previous analysis

for col, strategy in imputation_strategies.items():
    if col in ['building_age', 'year_built', 'land_size_m2', 'building_size_m2'] and grouping_columns:
        # Impute within groups if grouping columns are available
        if strategy == 'median':
            df[col] = df.groupby(grouping_columns)[col].transform(lambda x: x.fillna(x.median()))
        else: # strategy == 'mean'
            df[col] = df.groupby(grouping_columns)[col].transform(lambda x: x.fillna(x.mean()))
    else:
        # Impute without grouping
        if strategy == 'median':
            df[col] = df[col].fillna(df[col].median())
        else: # strategy == 'mean'
            df[col] = df[col].fillna(df[col].mean())

for col in ['building_age', 'year_built']:
    if col in imputation_strategies and imputation_strategies[col] == 'median':
        df[col] = df[col].fillna(df[col].median())
    elif col in imputation_strategies and imputation_strategies[col] == 'mean':
         df[col] = df[col].fillna(df[col].mean())

print("Missing values after imputation:")
print(df[imputation_strategies.keys()].isnull().sum())
```

```python
# Cek ulang missing value

imputed_columns = list(imputation_strategies.keys())
missing_values_after_imputation = df[imputed_columns].isnull().sum()
print("Missing values in imputed columns after imputation:")
print(missing_values_after_imputation)
```

```python
# Menghitung jumlah outlier yang ada

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate Q1, Q3, and IQR for each numerical column
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = {}
for col in numerical_cols:
    outliers[col] = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])]

print("Number of outliers detected using IQR:")
for col, outlier_df in outliers.items():
    print(f"'{col}': {len(outlier_df)} outliers")
```

```python
# Mengambil yang outliernya diatas 5% dari jumlah data untuk dilakukan binning.


outlier_percentage_threshold = 0.05
data_size = len(df)
outlier_count_threshold = data_size * outlier_percentage_threshold


columns_to_bin = [col for col, outlier_df in outliers.items() if len(outlier_df) > outlier_count_threshold]

print(f"Columns with more than {outlier_percentage_threshold*100}% outliers that will be binned:")
print(columns_to_bin)

# Binning sesuai dengan range
for col in columns_to_bin:
    binned_col_name = f'{col}_binned'

    if col == 'price_in_rp':
        # Apply custom binning for 'price_in_rp'
        bins = [0, 5e8, 1e9, 2e9, 5e9, df['price_in_rp'].max()]
        labels = ["Sangat Murah", "Murah", "Menengah", "Mahal", "Sangat Mahal"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    elif col == 'bedrooms':
        # Apply custom binning for 'bedrooms'
        bins = [0, 2, 4, 6, 20, df['bedrooms'].max()]
        labels = ["Kecil", "Sedang", "Besar", "Sangat Besar", "Ekstrem"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    elif col == 'land_size_m2':
        # Apply custom binning for 'land_size_m2'
        bins = [0, 72, 120, 200, 500, df['land_size_m2'].max()]
        labels = ["Kecil", "Sedang", "Luas", "Sangat Luas", "Ekstrem"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    elif col == 'building_age':
        # Apply custom binning for 'building_age'
        bins = [0, 5, 10, 20, 30, 50, df['building_age'].max()]
        labels = ["Baru", "Relatif Baru", "Sedang", "Cukup Tua", "Tua", "Sangat Tua"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    elif col == 'year_built':
        # Apply custom binning for 'year_built'
        bins = [df['year_built'].min(), 1945, 1970, 2000, 2020, df['year_built'].max()]
        labels = ["Sebelum 1945 (Bangunan Tua/Heritage)", "1946â€“1970 (Bangunan Lama)", "1971â€“2000 (Bangunan Sedang)", "2001â€“2020 (Bangunan Baru)", "2021+ (Pembangunan Terbaru)"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    elif col == 'building_size_m2':
        # Apply custom binning for 'building_size_m2'
        bins = [0, 36, 72, 150, 300, 1000, df['building_size_m2'].max()]
        labels = ["Sangat Kecil", "Kecil", "Sedang", "Besar", "Sangat Besar", "Ekstrem"]
        df[binned_col_name] = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True, right=True)
    else:
        # Apply equal-width binning for other columns
        num_bins = 10  # Using 10 bins as a starting point
        # pd.cut with labels=False will result in integer bins, need to create string labels
        bins = pd.cut(df[col], bins=num_bins, include_lowest=True, duplicates='drop')
        df[binned_col_name] = bins.astype(str) # Convert the interval labels to string


    print(f"\nValue counts for the binned column '{binned_col_name}':")
    print(df[binned_col_name].value_counts().sort_index())

# Display the first few rows with the new binned columns
print("\nDataFrame with new binned columns:")
display(df.head())
```

```python
# Distribusi Harga Rumah
# Histogram menunjukkan bahwa mayoritas harga rumah terkonsentrasi di kisaran harga
# murah menengah dan mahal, menunjukkan distribusi miring ke kanan dengan beberapa properti memiliki harga yang jauh lebih tinggi.
plt.figure(figsize=(10, 6))
sns.histplot(df['price_in_rp_binned'], kde=True)
plt.xlabel('Price in RP')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

```python
# Distribusi Harga Rumah Berdasarkan Kota
# Boxplot menunjukkan bahwa harga rumah bervariasi secara signifikan di setiap kota.
# Jakarta Pusat dan Jakarta Selatan umumnya memiliki harga median yang lebih tinggi dan rentang harga yang lebih luas dibandingkan kota-kota lain seperti Bekasi, Bogor, dan Depok.)
plt.figure(figsize=(12, 8))
sns.boxplot(x='city', y='price_in_rp_binned', data=df)
plt.xlabel('City')
plt.ylabel('Price in RP (Binned)')
plt.title('House Price Distribution by City')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Distribusi Fitur Kategoris (Barplot): Barplot menunjukkan bahwa 'rumah' adalah 'tipe_properti' yang dominan.
# Sebagian besar properti memiliki 1 atau 2 'lantai'.
# 'Kondisi_properti' sebagian besar 'bagus'.
# 'Perabotan' sebagian besar 'tanpa perabotan'.)

categorical_cols = ['property_type', 'floors', 'property_condition', 'furnishing']

plt.figure(figsize=(15, 10))

for i, col in enumerate(categorical_cols):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Count')
    if col in ['property_condition', 'furnishing']:
        plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Heatmap menunjukkan korelasi antar fitur numerik.Terdapat korelasi positif yang signifikan antara 'kamar tidur' dan 'kamar mandi', 'luas_lahan_m2' dan 'luas_bangunan_m2', serta 'kamar_tidur_pembantu' dan 'kamar_mandi_pembantu'. 'usia_bangunan' dan 'tahun_dibangun' memiliki korelasi negatif yang kuat.
numerical_cols = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# Relationship antara Land Size and Building Size
# Scatterplot menunjukkan hubungan antara ukuran tanah dan ukuran bangunan, yang mengindikasikan apakah properti dengan luas tanah yang lebih besar juga memiliki bangunan yang lebih besar.\
plt.figure(figsize=(10, 6))
sns.scatterplot(x='land_size_m2', y='building_size_m2', data=df)
plt.xlabel('Land Size (m^2)')
plt.ylabel('Building Size (m^2)')
plt.title('Relationship between Land Size and Building Size')
plt.show()
```
